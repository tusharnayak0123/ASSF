1. What is the difference between interpreted and compiled languages?
- The key difference between interpreted and compiled languages lies in how they are executed by the computer.

- Compiled Languages
The source code is translated into machine code by a compiler before execution.
The entire program is compiled at once, generating an executable file.
Runs faster since it is directly executed by the CPU.
Errors must be fixed before execution, as the program won’t run if there are compilation errors.
Examples: C, C++, Rust, Go
Interpreted Languages
The source code is executed line by line by an interpreter at runtime.
No separate compilation step; errors appear during execution.
Slower than compiled languages since the code is translated on the fly.
More flexible and easier to debug.
Examples: Python, JavaScript, PHP, Ruby
Hybrid Approach
Some languages use both compilation and interpretation:

- Java: Compiled into bytecode by the Java compiler, then interpreted by the Java Virtual Machine (JVM).
Python: Compiled into an intermediate bytecode (.pyc), which is then interpreted by the Python interpreter.

18. What is the difference between os.fork() and multiprocessing in Python ?
- Both os.fork() and the multiprocessing module are used for creating new processes in Python, but they have key differences in terms of portability, usability, and functionality.
-   os.fork() – Low-Level Process Creation (Unix/Linux Only)
  - Directly creates a child process by duplicating the parent process.
  -Returns 0 in the child process and the child’s PID in the parent process.
  -Only available on Unix/Linux (not supported on Windows).
  - Requires manual inter-process communication (IPC) using pipes, sockets, or shared memory.
- multiprocessing – High-Level Process Management (Cross-Platform )
  - Provides a portable way to create and manage processes (works on Windows, Linux, macOS).
  - Uses the Process class to create and manage multiple processes.
  - Built-in IPC mechanisms like queues, pipes, and shared memory.
  - Supports both process-based parallelism and thread-based parallelism.

```python
#Q4. Write a Python script that reads from one file and writes its content to another file?

def copy_file(source_file, destination_file):
    try:
        with open(source_file, "r") as src:
            content = src.read()

        with open(destination_file, "w") as dest:
            dest.write(content)

        print(f"Contents copied from '{source_file}' to '{destination_file}' successfully!")

    except FileNotFoundError:
        print(f"Error: The file '{source_file}' does not exist!")
    except Exception as e:
        print(f"An error occurred: {e}")

# Example Usage
source_file = "source.txt"
destination_file = "destination.txt"
copy_file(source_file, destination_file)
```

```python
#Q9. How can you read a file line by line and store its content in a list in Python?

def read_file_to_list(file_path):
    try:
        with open(file_path, "r") as file:
            lines = file.readlines()  # Read all lines into a list
        return [line.strip() for line in lines]  # Remove newline characters
    except FileNotFoundError:
        print(f"Error: The file '{file_path}' does not exist!")
        return []
    except Exception as e:
        print(f"An error occurred: {e}")
        return []

# Example Usage
file_path = "example.txt"  # Change this to your file name
lines_list = read_file_to_list(file_path)

# Print the list if not empty
if lines_list:
    print("File content as a list:", lines_list)

```

```python
#Q14. Write a program that uses the logging module to log both informational and error messages?

import logging

# Configure logging to write logs to a file and display in the console
logging.basicConfig(filename="app.log", level=logging.INFO,
                    format="%(asctime)s - %(levelname)s - %(message)s")

def divide_numbers(a, b):
    try:
        logging.info(f"Attempting to divide {a} by {b}")
        result = a / b
        logging.info(f"Division successful: {result}")
        return result
    except ZeroDivisionError:
        logging.error("Error: Division by zero is not allowed!")
        return None
    except Exception as e:
        logging.error(f"Unexpected error occurred: {e}")
        return None

# Example usage
divide_numbers(10, 2)  # Successful case
divide_numbers(10, 0)  # Error case

print("Log messages have been recorded in 'app.log'.")
```

```python
#Q18. How would you implement a basic logging setup that logs to a file with rotation after 1MB?

import logging
from logging.handlers import RotatingFileHandler

# Configure logging
log_file = "app.log"

# Set up rotating file handler (Max size: 1MB, Keep 3 backup files)
handler = RotatingFileHandler(log_file, maxBytes=1*1024*1024, backupCount=3)

# Log format
formatter = logging.Formatter("%(asctime)s - %(levelname)s - %(message)s")
handler.setFormatter(formatter)

# Logger setup
logger = logging.getLogger("RotatingLogger")
logger.setLevel(logging.INFO)
logger.addHandler(handler)

# Example log messages
for i in range(10000):  # Generate many log messages to test rotation
    logger.info(f"Log message {i}")

print(f"Logging setup complete. Logs are stored in {log_file}")
```

```python
#Q19. Write a program that handles both IndexError and KeyError using a try-except block?

def handle_errors():
    my_list = [10, 20, 30]
    my_dict = {"a": 1, "b": 2}

    try:
        # Trying to access an out-of-range index (IndexError)
        print("List element:", my_list[5])

        # Trying to access a non-existent key in dictionary (KeyError)
        print("Dictionary value:", my_dict["c"])

    except IndexError:
        print("Error: List index is out of range!")

    except KeyError:
        print("Error: Dictionary key not found!")

handle_errors()
```

```python
#Q21. Write a Python program that reads a file and prints the number of occurrences of a specific word?
def count_word_occurrences(file_path, target_word):
    try:
        with open(file_path, "r", encoding="utf-8") as file:
            content = file.read().lower()  # Read and convert to lowercase for case-insensitive search
            word_count = content.split().count(target_word.lower())  # Count occurrences
        print(f"The word '{target_word}' appears {word_count} times in {file_path}.")
    except FileNotFoundError:
        print("Error: The file does not exist!")
    except Exception as e:
        print(f"An error occurred: {e}")

# Example usage
file_path = "example.txt"  # Change to your actual file name
target_word = "Python"  # Change to the word you want to count
count_word_occurrences(file_path, target_word)
```

```python
#Q22. How can you check if a file is empty before attempting to read its contents?

import os

def is_file_empty(file_path):
    return os.stat(file_path).st_size == 0  # Check if file size is 0 bytes

file_path = "example.txt"  # Change this to your file

if os.path.exists(file_path):
    if is_file_empty(file_path):
        print(f"The file '{file_path}' is empty!")
    else:
        with open(file_path, "r") as file:
            content = file.read()
            print("File Content:\n", content)
else:
    print("Error: The file does not exist!")
```
